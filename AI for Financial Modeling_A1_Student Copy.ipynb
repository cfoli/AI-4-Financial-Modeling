```python
# Before we begin, we'll import the necessary packages
import matplotlib
import matplotlib.pyplot as plt

# We will also use a 'magic method' to tell our plotting package, matplotlib, to print our plots
# in our notebook and to save the outputs for future use

%matplotlib inline
```

- Here are two things that you might want to remember:

    - ***Axes*** represents a plot
    
        - Do not confuse axes with x-axis/y-axis. In fact, each plot (i.e., axes) will have its own x-axis and y-axis
        
    - ***Figure*** is the final image; it can contain one or more axes (plots)

- We start by creating a new figure object

    - We can then attach axes (or plots) to the above figure

Quick Description of what 'inline' means:

With this backend, the output of plotting commands is displayed inline within frontends like the Jupyter notebook, directly below the code cell that produced it. The resulting plots will then also be stored in the notebook document.

```python
# Create a figure object and generate a 2x2 axes/subplots
# Notice how the resultant figure is blank before we populate it with data
fig = plt.figure()
ax = fig.subplots(nrows=2, ncols=2)
```

```python
# Let us check the dimennsions of ax
ax.shape
```

```python
# An alternate way to create fig and ax
fig,ax = plt.subplots(nrows=2,ncols=2)
```

- Since ax is an array, we can assign plots individually to the array

```python
# Assign the histogram of Pclass to the first axes
ax[0,0] = titanic_data['Pclass'].hist()
```

```python
# Assign the histogram of Age to the second axes.... and so on
ax[0,1] = titanic_data['Age'].hist()
```

```python
ax[1,0] = titanic_data['SibSp'].hist()
```

```python
ax[1,1] = titanic_data['Fare'].hist()
```

```python
# Instead of plotting each feature one-by-one, we can plot all histograms together at one time
fig, myaxes = plt.subplots(nrows=2, ncols=2)
titanic_data[['Age','Pclass','SibSp','Fare']].hist(ax=myaxes)
plt.show()
```

- Note that we do not have to explicitly assign ax[0,0], ax[0,1], ax[1,0] or ax[1,1]
- There is another way to create a 2x2 grid of subplots (which provides more flexibility)
    - In particular, we do not need to specify the size of ax matrix

```python
# Compared to the earlier version, this version has a distinct advantage:
# we do not have to specify the dimensions of the subplot.
fig, myaxes = plt.subplots()
titanic_data[['Age','Pclass','SibSp','Fare']].hist(ax=myaxes)
plt.show()
```

- The above function call yields a warning (which we will ignore for now)

- Note that we do not specify the ***exact dimension of subplots*** we need in the subplots() call

    - The usefulness of this alternate syntax will be evident in the next example

```python
# Make a call to subplots with no fixed dimensions
fig,myaxes = plt.subplots()

# Next, we will increase the height [to 50] and width [to 30] of the figure
fig.set_figheight(40)
fig.set_figwidth(20)

# Finally, plot the histograms
titanic_data.hist(ax=myaxes)
plt.show()
```
